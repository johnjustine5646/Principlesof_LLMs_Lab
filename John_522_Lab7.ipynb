# Implement tokenization and encoding for any Domain-specific dataset

```python
import pandas as pd
from transformers import DistilBertTokenizer
import re

df = pd.read_csv('/content/spotify_millsongdata.csv')
df.head()
```

```python
# Text cleaning function
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove non-alphabetic characters
    text = text.lower()  # Convert to lowercase
    return text

# Clean the lyrics
df['clean_text'] = df['text'].apply(clean_text)
```

function takes a list of texts and the tokenizer, and returns encoded tensors with padding and truncation applied.

```python
# Load pre-trained tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

# Tokenize and encode
def tokenize_and_encode(text_list, tokenizer):
    encoding = tokenizer(text_list, truncation=True, padding=True, return_tensors='pt')
    return encoding

encoded_data = tokenize_and_encode(df['clean_text'].tolist(), tokenizer)

# Check encoded data
print("Input IDs:")
print(encoded_data['input_ids'])

print("\nAttention Mask:")
print(encoded_data['attention_mask'])
```
